squared bias: As the flexibility of the model increases, predictions that are more similar to the training values can be obtained, so the squared bias decreases.

# 10. Mini-batch gradient descent (20pt)

**In the Lab 2, we learned how to implement gradient descent and stochastic gradient descent. For this question, implement mini-batch gradient descent. Set the ````batch_size```` to 10.**

`````python
from numpy.linalg import inv

boston = pd.read_csv('data/Boston.csv') # load data
#boston = pd.read_csv('/content/drive/MyDrive/ML IE/HW1/data/Boston.csv')

X_train = boston[["lstat"]].values # regression using all input columns
X_train = sm.add_constant(X_train)  #add_constant  
y_train = boston['medv'].values

beta_direct = inv(X_train.T.dot(X_train)).dot(X_train.T).dot(y_train)
print(beta_direct)

beta = np.random.rand(2)
alpha = 0.001  # The learning Rate
epochs = 1000  # The number of iterations to perform gradient descent
n = float(len(X_train)) # Number of elements in X
batch_size = 10

######################## Write your answere here ########################
n_batches = int(n // batch_size)
cost = []
for epoch in range(epochs):
    temp = 0
    for i in range(n_batches):
        y_pred = np.dot(X_train[i*batch_size:(i+1)*batch_size], beta)
        error = y_pred - y_train[i*batch_size:(i+1)*batch_size]    
        dbeta = (1/batch_size) * np.dot(X_train[i*batch_size:(i+1)*batch_size].T, error)
        beta = beta - alpha * dbeta

        temp += np.dot(error.T, error)
        
    J = (1/n) * temp
    cost.append(J)

#########################################################################

print("Learned parameters through SGD: {}".format(beta))
print("Directly learned parameters: {}".format(beta_direct))   
plt.plot(np.arange(epochs), cost)
plt.title("Learning curve")
plt.xlabel("Epoch")
plt.ylabel(r"J($\beta$)");
`````
